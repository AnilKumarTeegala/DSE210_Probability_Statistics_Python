```python
import numpy as np
import sys
import os

def read_data(path, file_name):
    with open(os.path.join(path, file_name), 'r') as f:
        data_dic = {}
        increment = 0
        for line in f:
            doc_word_count_gen = (int(i) for i in line.split()) # generator comprehension
            doc_id = next(doc_word_count_gen)
            word_id = next(doc_word_count_gen)
            count = next(doc_word_count_gen)
            data_dic[increment] =  [doc_id, word_id, count] 
            increment +=1
    return data_dic

#path = '20news-bydate/matlab/'
# data_dic = read_data(path, 'train.data')
print data_dic[0]
```

```python
def read_label(path, filename):
    with open(os.path.join(path, 'train.label'), 'r') as f:
        label_dic = {0:0}
        pi = {k:0 for k in xrange(21)}
        increment = 1
        for line in f:
            label_dic[increment] = int(line)
            increment += 1
            pi[int(line)] += 1
        for key in pi:    
            pi[key] = [float(pi[key])/len(label_dic)]
        pi[0] = [1.0]
        for key in pi:    
            pi[key] = np.log2(pi[key])[0]
    return label_dic, pi
path = '20news-bydate/matlab/'
label_dic, pi = read_label(path, 'train.label')
label_dic
```

```python
def setup_multinomial_model(label, data):
    m = np.zeros((21, 61189))
    len_data = len(data)

    for i in xrange(len_data):
        doc_id = data[i][0]
        word_id = data[i][1]
        count = data[i][2]
        classify = label[doc_id]
        m[classify][word_id] += count
    # Remove stop words
    stop_word = {12:"of", 23:"and",139:"an",978:"am",297:"at",51:"but",52:"with",33:"to",48:"on",27:"are",29:"the",72:"can",1367:"else",81:"for",301:"he",389:"she",99:"so", 978:"am", 30:"in"}
    for k in stop_word:
        m[:, k] = 0.0

    m += 1 # smoothing of the data
    m[:,0] = 0.0

    s = np.sum(m, axis = 1) #summing across the rows ie classes
    s_trans = np.transpose([s]) #Transposing a 1-D array returns an unchanged view of the original array.

    m = m / s_trans
    m[:,0] = 1.0
    m = np.log2(m)
    return m
```

```python
path = '20news-bydate/matlab/'
label_array, pi = read_label(path, 'train.label')
data_array = read_data(path, 'train.data')
m = setup_multinomial_model(label_array, data_array)
test_label, _ = read_label(path, 'test.label')
test_data = read_data(path, 'test.data')

len_test_data = len(test_data)
number_doc_plus_1 = len(test_label)

test_m = np.zeros((number_doc_plus_1, 61189))
for i in xrange(len_test_data):
    doc_id = test_data[i][0]
    word_id = test_data[i][1]
    count = test_data[i][2]
    test_m[doc_id][word_id] += count

test_m = np.log2(1+test_m)
print test_m
error = 0
for i in xrange(1, number_doc_plus_1):
    if i > 4:
        break
    cur_doc = np.transpose(test_m[i])
    cur_s = np.sum(cur_doc * m, axis = 1)
    final = cur_s + pi
    final = final[1:]
    label = np.argmax(final) + 1
    if label != test_label[i]:
        error += 1
error * 100.0 / (number_doc_plus_1 - 1)
```
