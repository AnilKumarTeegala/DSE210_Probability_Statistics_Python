```python
%cd data/Animals_with_Attributes/
new_x = np.fromfile("predicate-matrix-continuous.txt", sep=' ')
%cd ../..
```

```python
%cd data/Animals_with_Attributes/

df = pd.read_csv('classes.txt', sep='\t', header=None) #,sep='\t'

#c = np.genfromtxt("classes.txt")
classes = np.array(df)
#np.loadtxt(c, delimiter=','
%cd ../..
```

```python
%cd data/Animals_with_Attributes/
df = pd.read_csv('predicates.txt', sep='\t', header=None) #,sep='\t'
predicates = np.array(df)
%cd ../..
```

```python
#predicates
```

```python
kmeans = KMeans(10, init='random')
```

```python
fit = kmeans.fit(new_x)
```

```python
mypreds = fit.predict(new_x)
```

```python
classes
```

```python
mypreds
```

```python
mypreds==1
```

```python
classes[mypreds==1]
```

```python
result = {}
for i in range(10):
    result[i] = classes[mypreds==i]
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
l = linkage(new_x, method='ward')
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 10, 15
#plt.title('Hierarchical Clustering Dendrogram')
#plt.xlabel('sample index')
#plt.ylabel('distance')
dendrogram(l, orientation='right', labels=classes[:,1])
#plt.figure(figsize=(70,10))

plt.show()
```
